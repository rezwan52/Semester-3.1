#3.Sum of Digits

Write a program that takes an integer input and calculates the sum of its digits using a while loop.

```python
def sum_of_digits(n):
    if n < 0:
        print("Please enter a positive integer.")
        return
    total_sum = 0

    while n > 0:
        digit = n % 10
        total_sum += digit
        n //= 10

    return total_sum

num = int(input("Enter an integer: "))

result = sum_of_digits(num)

print(f"The sum of the digits of {num} is: {result}")
```

#4.Prime Number Generator

Implement a program that generates and prints the first 10 prime numbers using a loop.

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def generate_first_n_primes(n):
    primes = []
    num = 2
    while len(primes) < n:
        if is_prime(num):
            primes.append(num)
        num += 1
    return primes

first_10_primes = generate_first_n_primes(10)
print("The first 10 prime numbers are:", first_10_primes)
```

#5.Palindrome Check

Create a program that checks if a given word is a palindrome using a loop.

```python
def is_palindrome(word):
    word = word.lower()

    left, right = 0, len(word) - 1
    while left < right:
        if word[left] != word[right]:
            return False
        left += 1
        right -= 1

    return True
input_word = input("Enter a word: ")

if is_palindrome(input_word):
    print(f"{input_word} is a palindrome.")
else:
    print(f"{input_word} is not a palindrome.")
```

#5.Dictionary Key Removal

Write a Python program that takes a dictionary and a key as input, removes the key-value pair with the specified key, and prints the modified dictionary.


```python
def remove_key(dictionary, key_to_remove):
    if key_to_remove in dictionary:
        del dictionary[key_to_remove]
    return dictionary

my_dict = {'apple': 'red', 'banana': 'yellow', 'grape': 'purple'}
key_to_remove = 'banana'
result = remove_key(my_dict, key_to_remove)
print(f"Dictionary after removing key '{key_to_remove}':", result)
```

#5.Function List Element Count

Write a function that takes a list as input and returns a dictionary containing the count of each unique element in the list.

```python
def element_count(input_list):
    count_dict = {}
    for elem in input_list:
        count_dict[elem] = count_dict.get(elem, 0) + 1
    return count_dict

result = element_count([1, 2, 2, 3, 3, 3, 4, 4, 4, 4])
print("Element count:", result)
```
